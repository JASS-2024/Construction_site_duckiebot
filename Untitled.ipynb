```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.cvtColor(cv2.imread("/home/ilia/Изображения/Снимок экрана от 2024-03-26 18-15-55.png"), cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

```python
hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

# Определение диапазона зеленого цвета в HSV
lower_green = np.array([80, 40, 40])  # Нижний порог зеленого цвета
upper_green = np.array([100, 255, 255]) # Верхний порог зеленого цвета

# Создание маски для зеленого цвета
mask = cv2.inRange(hsv_image, lower_green, upper_green)

# Применение маски к исходному изображению
green_areas = cv2.bitwise_and(image, image, mask=mask)

# Отображение только зеленых областей
plt.imshow( green_areas)
```
